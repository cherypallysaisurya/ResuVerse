###Parsing the Resume Text.

```python
import io
import re
from pdfminer.high_level import extract_text_to_fp
from pdfminer.layout import LAParams

def extract_text_from_pdf(pdf_path):
    """
    Extract text from a PDF file.

    :param pdf_path: Path to the PDF file.
    :return: Extracted text as a string.
    """
    output_string = io.StringIO()
    with open(pdf_path, 'rb') as f:
        extract_text_to_fp(f, output_string, laparams=LAParams(), output_type='text', codec=None)
    return output_string.getvalue()

def parse_resume_text(text):
    """
    Parse the resume text to extract key details like name, email, phone number, and skills.

    :param text: The extracted resume text.
    :return: A dictionary with parsed details.
    """
    # Assume the first non-empty line is the candidate's name.
    lines = [line.strip() for line in text.splitlines() if line.strip()]
    name = lines[0] if lines else "Name not found"

    # Use regex to find an email address.
    email_match = re.search(r'\S+@\S+', text)
    email = email_match.group(0) if email_match else "Email not found"

    # Use regex to find a phone number pattern (e.g., (561) 248-2079)
    phone_match = re.search(r'\(\d{3}\)\s*\d{3}[-\s]?\d{4}', text)
    phone = phone_match.group(0) if phone_match else "Phone number not found"

    # Extract the 'Skills' section using a regex that captures everything after "Skills"
    skills = ""
    skills_pattern = re.compile(r'Skills\s*\n([\s\S]+)', re.IGNORECASE)
    skills_match = skills_pattern.search(text)
    if skills_match:
        # Capture everything after the "Skills" header and clean it up.
        skills = skills_match.group(1).strip()
    else:
        skills = "Skills not found"

    return {
        "name": name,
        "email": email,
        "phone": phone,
        "skills": skills
    }

if __name__ == "__main__":
    # Replace with the actual path to your resume PDF.
    pdf_file_path = "/content/saisurya_ta_resume.pdf"

    # Step 1: Extract text dynamically from the PDF.
    extracted_text = extract_text_from_pdf(pdf_file_path)
    #print("Extracted Text:\n", extracted_text)

    # Step 2: Parse the dynamically extracted text.
    parsed_resume = parse_resume_text(extracted_text)
    print("\nParsed Resume Details:")
    for key, value in parsed_resume.items():
        print(f"{key.capitalize()}: {value}")
```

```python
import io
import re
from pdfminer.high_level import extract_text_to_fp
from pdfminer.layout import LAParams
from sentence_transformers import SentenceTransformer, util

def extract_text_from_pdf(pdf_path):
    """
    Extract text from a PDF file.

    :param pdf_path: Path to the PDF file.
    :return: Extracted text as a string.
    """
    output_string = io.StringIO()
    with open(pdf_path, 'rb') as f:
        extract_text_to_fp(f, output_string, laparams=LAParams(), output_type='text', codec=None)
    return output_string.getvalue()

def parse_resume_text(text):
    """
    Parse the resume text to extract key details like name, email, phone number, and skills.

    :param text: The extracted resume text.
    :return: A dictionary with parsed details.
    """
    # Assume the first non-empty line is the candidate's name.
    lines = [line.strip() for line in text.splitlines() if line.strip()]
    name = lines[0] if lines else "Name not found"

    # Use regex to find an email address.
    email_match = re.search(r'\S+@\S+', text)
    email = email_match.group(0) if email_match else "Email not found"

    # Use regex to find a phone number pattern (e.g., (561) 248-2079)
    phone_match = re.search(r'\(\d{3}\)\s*\d{3}[-\s]?\d{4}', text)
    phone = phone_match.group(0) if phone_match else "Phone number not found"

    # Extract the 'Skills' section using a regex that captures everything after "Skills"
    skills = ""
    skills_pattern = re.compile(r'Skills\s*\n([\s\S]+)', re.IGNORECASE)
    skills_match = skills_pattern.search(text)
    if skills_match:
        skills = skills_match.group(1).strip()
    else:
        skills = "Skills not found"

    return {
        "name": name,
        "email": email,
        "phone": phone,
        "skills": skills
    }

def get_embedding(text, model):
    """
    Generate an embedding for the given text.

    :param text: Input text.
    :param model: The SentenceTransformer model.
    :return: Embedding vector.
    """
    return model.encode(text, convert_to_tensor=True)

if __name__ == "__main__":
    # Path to the resume PDF (update as needed)
    pdf_file_path = "/content/resumes/Resume.pdf"

    # Step 1: Extract text dynamically from the PDF.
    extracted_text = extract_text_from_pdf(pdf_file_path)
    print("Extracted Text:\n", extracted_text)

    # Step 2: Parse the extracted text.
    parsed_resume = parse_resume_text(extracted_text)
    print("\nParsed Resume Details:")
    for key, value in parsed_resume.items():
        print(f"{key.capitalize()}: {value}")

    # Step 3: Combine parsed resume details into one text string for embedding.
    resume_combined_text = (
        f"{parsed_resume['name']}\n"
        f"{parsed_resume['email']}\n"
        f"{parsed_resume['phone']}\n"
        f"Skills: {parsed_resume['skills']}"
    )

    # Initialize the SentenceTransformer model.
    model = SentenceTransformer('all-MiniLM-L6-v2')

    # Generate the embedding for the resume from dynamic data.
    resume_embedding = get_embedding(resume_combined_text, model)
    print("\nGenerated resume embedding with shape:", resume_embedding.shape)

    # Get job description dynamically. Here you could take it from user input.
    # For demonstration, we simulate dynamic input:
    job_description = input("Enter the job description: ")

    # Generate the embedding for the job description.
    job_embedding = get_embedding(job_description, model)
    print("Generated job description embedding with shape:", job_embedding.shape)

    # Compute cosine similarity between the resume and the job description.
    similarity = util.pytorch_cos_sim(resume_embedding, job_embedding)
    print("Cosine Similarity between resume and job description:", similarity.item())
```
